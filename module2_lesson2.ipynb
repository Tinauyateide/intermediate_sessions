```python
import pandas as pd
import sqlite3
import sqlalchemy
from sqlalchemy import create_engine

pg_url = 'postgresql+psycopg2://postgres:postgres@localhost:5432/demo'

engine = create_engine(pg_url, pool_recycle=-1)

print('Engine created successfully')

df = pd.read_csv("/home/tina/intermidiate_sessions/udemy_courses.csv")

df.to_sql('Udemy_Course', con=engine, index=False)

print("Done"),
```

```python
# define a query function
def query_db(query: str, db_conn: sqlalchemy.engine.base.Connection) -> pd.DataFrame:
    df = pd.read_sql(query, db_conn)
    #print(df)
    return df

query_db("SELECT * FROM my_udemy_course LIMIT 5", engine)
```

```python
query_db("""select course_title, num_subscribers from My_Udemy_Course where course_id = (
select course_id from My_Udemy_Course
    where EXTRACT(MONTH FROM published_timestamp::date) = 1 
    order by num_subscribers desc
    limit 1
)
""", engine)            # 1 at end of the where clause for published _time stamp signifies the first month in other to get a different month you can input the month num
```
